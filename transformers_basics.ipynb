```python
from transformers import pipeline


classifier = pipeline("sentiment-analysis")


result = classifier("I love learning machine learning!")
print(result)
```

```python
sentences = [
    "This movie is terrible!",
    "I had a great day at work.",
    "The weather is okay.",
    "I'm feeling sad today.",
    "Best experience ever!"
]

for sentence in sentences:
    result = classifier(sentence)
    print(f"{sentence}")
    print(f"  → {result[0]['label']} ({result[0]['score']:.4f})")
    print()
```

```python
medical_texts = [
    "The patient shows signs of improvement after treatment.",
    "Severe complications observed during surgery.",
    "Patient is stable and recovering well.",
    "Critical condition, immediate intervention required."
]

for text in medical_texts:
    result = classifier(text)
    print(f"{text}")
    print(f"  → {result[0]['label']} ({result[0]['score']:.4f})")
    print()
```

```python
qa = pipeline("question-answering")

context = """
The patient is a 45-year-old male with a history of hypertension.
He presented with chest pain and shortness of breath.
ECG showed signs of myocardial infarction.
"""

question = "What is the patient's age?"

answer = qa(question=question, context=context)
print(f"Soru: {question}")
print(f"Cevap: {answer['answer']}")
print(f"Confidence: {answer['score']:.4f}")
```

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    model="dbmdz/bert-large-cased-finetuned-conll03-english",
    aggregation_strategy="simple"
)

text = "The patient John Smith was treated at Mayo Clinic in Rochester."

entities = ner(text)

for entity in entities:
    print(f"{entity['word']}: {entity['entity_group']} ({entity['score']:.4f})")
```

```python

from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

model_name = "bvanaken/clinical-assertion-negation-bert"


tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)


medical_classifier = pipeline(
    "text-classification",
    model=model,
    tokenizer=tokenizer
)


result = medical_classifier("The patient denies any chest pain.")
print(result)
```

```python
print(medical_classifier("The patient has severe chest pain."))
```
